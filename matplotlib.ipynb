```python
#lets try another dataset
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
#create a histogram of age
heart_disease["age"].plot.hist()
```

```python
heart_disease

```

```python
heart_disease.plot.hist(figsize=(10,30),subplots=True);
```

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target");
```

```python
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
             x='age',
             y='chol',
             c='target',
             ax=ax);
# ax.set_xlim([45,100]);
```

```python
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
             x='age',
             y='chol',
             c='target',
             ax=ax);
ax.set_xlim([45,100]);
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

#plot the data
scatter= ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);
#customize the plot
ax.set(title="Heart Disease and Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

#add a legend
ax.legend(*scatter.legend_elements(), title="Target");

#add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--')
```

```python
over_50.head()
```

```python
talach_145=heart_disease[heart_disease["thalach"] > 145]
talach_145.head()
```

```python
fig, (ax1 , ax2)= plt.subplots(figsize= (10,10),
                                  nrows=2);
scatter=ax1.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

ax1.set(title="Heart Disease",
       xlabel="Age",
       ylabel="Cholesterol");
ax1.legend(*scatter.legend_elements(), title="Target")
ax1.axhline(y=over_50["chol"].mean(),
            linestyle='--')

scatter=ax2.scatter(x=talach_145["age"],
                    y=talach_145['chol'],
                    c=talach_145['target']);
ax2.set(title= "Heart Rate",
        xlabel="Age hola",
        ylabel="Jhos");
ax2.legend(*scatter.legend_elements(), title = "Talach")
ax2.axhline(y=talach_145["chol"].mean(),
            linestyle='--');
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");

plt.style.use('seaborn-v0_8');
```
